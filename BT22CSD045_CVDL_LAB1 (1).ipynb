```python
# Read and Display an Image
import cv2 as cv
from google.colab.patches import cv2_imshow

img = cv.imread('dog.jpg')
cv2_imshow(img)

```

```python
# Draw a Circle on an Image
circle_img = img.copy()
HI_IMG=cv.circle(circle_img, (85, 95), 50, (255, 0, 0), 5)
cv2_imshow(HI_IMG)

```

```python
# Draw a Line on an Image
line_img = img.copy()
cv.line(line_img, (50, 50), (240, 230), (0, 255, 0), 3)
cv2_imshow(line_img)

```

```python
# Draw a Rectangle on an Image
rect_img = img.copy()
cv.rectangle(rect_img, (30, 30), (100, 100), (0, 0, 255), 3)
cv2_imshow(rect_img)

```

```python
# Draw Multiple Shapes on an Image
multi_shape_img = img.copy()
cv.circle(multi_shape_img, (100, 100), 40, (255, 0, 0), 5)
cv.line(multi_shape_img, (100, 50), (250, 80), (0, 255, 0), 2)
cv.rectangle(multi_shape_img, (30, 50), (100, 100), (0, 0, 255), 3)
cv2_imshow(multi_shape_img)

```

```python
# Display Multiple Images in Different Windows
top_left = img[0:150, 0:150]
top_right = img[0:150, 150:300]
cv2_imshow(top_left)
cv2_imshow(top_right)

```

```python
# Apply Filters and Display Results
gaussian_blur = cv.GaussianBlur(img, (15, 15), 0)
cv2_imshow(gaussian_blur)

```

```python
# Convert Image to Grayscale
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_img)


```

```python
# Resize an Image
resized_img = cv.resize(img, (150, 500))
cv2_imshow(resized_img)

```

```python
# Rotate an Image
(h, w) = img.shape[:2]
center = (w // 2, h // 2)
rotation_matrix = cv.getRotationMatrix2D(center, 65, 1.0)
rotated_img = cv.warpAffine(img, rotation_matrix, (w, h))
cv2_imshow(rotated_img)

```

```python
# Save an Image
cv.imwrite('modified_dog.jpg', rotated_img)

```

```python
# Display Image in Multiple Color Channels
blue, green, red = cv.split(img)
cv2_imshow(blue)
cv2_imshow(green)
cv2_imshow(red)

```

```python
# Draw Text on an Image
text_img = img.copy()
cv.putText(text_img, 'its my dog ', (50, 50), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
cv2_imshow(text_img)

```

```python
# Crop an Image
cropped_img = img[50:250, 50:250]
cv2_imshow(cropped_img)

```

```python
# Create a Blank Image and Draw Shapes
blank_img = cv.imread('dog.jpg')
blank_img[:] = (0, 0, 0)
cv.circle(blank_img, (200, 100), 50, (255, 0, 0), -1)
cv.rectangle(blank_img, (100, 100), (150, 150), (0, 255, 0), 2)
cv.line(blank_img, (100, 100), (100, 300), (0, 0, 255), 3)
cv2_imshow(blank_img)

```

```python
# Change Image Brightness and Contrast
bright_contrast_img = cv.convertScaleAbs(img, alpha=1.2, beta=50)
cv2_imshow(bright_contrast_img)
```
